```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
df.head()
```

```python
#now we will normalize then numerical data of out features , logistic regression is sensitive to the scalre of the features value
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Normalize the data
scaler = StandardScaler()
scaler.fit(df.drop('diagnosis', axis=1))
scaled_features = scaler.transform(df.drop('diagnosis', axis=1))

# Create the dataframe
df_feat = pd.DataFrame(scaled_features, columns=df.columns[:-1])

# Create the X and y variables
X = df_feat
y = df['diagnosis']
```

```python
#satndardScaler is sensitive to outlier bc it uses mean and standard deviation
```

```python
df_feat.head()
```

```python
#quick definition , standard deviates calculate how much the data points deviate from the mean value 
#low standard deviation: data is closely clustered around the mean and vice versa 
```

```python
from sklearn.linear_model import LogisticRegression

# Create the model
logmodel = LogisticRegression()
logmodel.fit(X, y)
```

```python
from sklearn.model_selection import train_test_split


# Split the dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# Test the model
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report

# Print the report
print(classification_report(y_test, predictions))
```

```python
#accuracy : correct predicitons / all predictions , basically how often the model is right 
#precision : TP / TP + FS , how often the positive predictions are correct 
# recall : TP / TP + FN , can a model predict all instance of positive classes 
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'penalty': ['l1', 'l2'],  # Type of regularization
    'solver': ['liblinear']  # 'liblinear' works with l1 and l2 penalties
}
```

```python
grid = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)

#Regularization helps control model complexity by discouraging overly large weights (coefficients) that might make the model too dependent on training data. This reduces overfitting and improves generalization.
#Without regularization, the model might memorize the training data rather than learning patterns that generalize well.
```
